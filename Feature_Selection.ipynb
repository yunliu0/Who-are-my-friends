```python
import pandas as pd
import pickle 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('samples_60000.txt', 'rb') as file:
    train = pickle.load(file)
```

```python
train
```

## Correlation Among all Features

```python
XX=pd.DataFrame()
```

```python
X=train.iloc[:,1:]
Y=train.iloc[:,0]
```

```python
train_corr = train.corr().abs()
train_corr
```

From above correlation matrix, we notice that the correlation among `Centrality Feature` and `node Features` are more than 0.9, so we only need one of these two feature class. Since the correlation for `Centrality Feature`  and  `Y` are larger than `node Features` with `Y`. we remove all the `node Features` except `shortest_path` from the dataset

```python
XX=X.iloc[:,4:]
XX.iloc[:5,:]
```

## Correlations Among _Similarity Features_

```python
Similarity_X=XX.iloc[:,1:7]
Similarity_X['Y']=Y
SimilarityX_corr = Similarity_X.corr().abs()
SimilarityX_corr
```

Since $corr$(`AAI`,`CN`)=0.987, and $corr$(`Y`,`CN`)=0.452 < $corr$(`AAI`,`Y`)=0.46580, we remove the feature `CN`.

```python
XX=XX.drop(['CN'],axis=1)
XX.iloc[:5,:]
```

## Correlations Among _Centrality Feature_

```python
Centrality_X=XX.iloc[:,7:]
Centrality_X['Y']=Y
CentralityX_corr = Centrality_X.corr().abs()
CentralityX_corr
```

```python
XX_cor = XX.corr().abs()
XX_cor
```
